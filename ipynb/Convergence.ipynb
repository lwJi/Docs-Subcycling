# Convergence Test for Different Subcycling Method (1D Linear Wave)

```julia
using Plots
```

```julia
include("./misc.jl")
```

```julia
function analytical_psi(t, x)
    return sin(2 * pi * (x - t))
end
```

```julia
function plot_error(dirs; nbuf = 2, lev = 1, tmax = 1.0, conv_order = 4,
                    parent_dir = expanduser("~/docker-workspace/simulations/"))
    fulldirs = parent_dir .* dirs
    dats = Misc.load_data(fulldirs, nbuf, lev, tmax)
    plt = plot()
    for i in 1:length(dats)
        x = dats[i][1]
        psi = dats[i][2]
        err = (psi - analytical_psi.(tmax, x))
        plt = plot!(x, err * (2^conv_order)^(i-1))
    end
    return plt
end
```

## 2 Levels with Mongwane's Subcycling + Transition Zone (4th order convergent)

```julia
plot_error(
    [
        "scalarwave_2levels_nx100_MongwaneTrans_ords3",
        "scalarwave_2levels_nx200_MongwaneTrans_ords3",
        "scalarwave_2levels_nx400_MongwaneTrans_ords3"
    ]; nbuf = 2, lev = 1, conv_order = 4, tmax = 0.8)
```
