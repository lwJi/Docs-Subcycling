# Explicit Runge Kutta Method

$$
\begin{align*}
Y_s&=y_n+h \Sigma_{i=1}^{s-1}a_{si}f(Y_i, t_n+hc_i)\\
y_{n+1}&=y_n+h \Sigma_{i=1}^{s}b_if(Y_i, t_n+hc_i)
\end{align*}
$$

or in terms of $k_i$s:
$$
\begin{align*}
k_i=&hf(Y_i,t_n+h c_i) \\
y_{n+1}=&y_n+\Sigma_{i=1}^{s}b_ik_i
\end{align*}
$$

populer RK4:
$$
\begin{align*}
&c_1=0, c_2=c_3=1/2, c_4=1 \\
&a_{21}=a_{32}=1/2, a_{31}=a_{41}=a_{42}=0, a_{43}=1\\
&b_{1}=b_{4}=1/6, b_{2}=b_{3}=1/3
\end{align*}
$$

#### Autonomous Assumption

$$
\begin{align*}
y'&=f, \\
y''&=f f_y, \\
y'''&=f (f_y)^2+f^2f_{yy}
\end{align*}
$$
