## WaveToyX

```julia
conv_order = 4
dirs = [
    "ConvWaveToy/gaussianWaveToy_64/output-0000/gaussianWaveToy_64",
    "ConvWaveToy/gaussianWaveToy_128/output-0000/gaussianWaveToy_128",
    "ConvWaveToy/gaussianWaveToy_256/output-0000/gaussianWaveToy_256",
]
parent_dir = expanduser("~/docker-workspace/simulations/")
fulldirs = parent_dir .* dirs
dats = Misc.load_data(fulldirs, 0, 1, 1.0; thorn="wavetoyx")
plt = plot()
for i in 1:length(dats)
    x = dats[i][1]
    err = abs.(dats[i][2])
    plt = plot!(x, err * (2^conv_order)^(i-1))
end
return plt
```
