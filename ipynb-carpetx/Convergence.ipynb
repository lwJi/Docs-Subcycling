# Convergence Test for Different Subcycling Method (1D Linear Wave)

```julia
using Plots
include("./misc.jl")
```

## ConvDIY

```julia
conv_order = 4
dirs = [
    "ConvDIY/gaussianDIY_64/output-0000/gaussianDIY_64",
    "ConvDIY/gaussianDIY_128/output-0000/gaussianDIY_128",
]
conv_bases = [1, 2, 4]
parent_dir = expanduser("~/docker-workspace/simulations/")
fulldirs = parent_dir .* dirs
dats = Misc.load_data(fulldirs, 0, 1, 1.0, thorn="testsubcyclingmc")
plt = plot()
for i in 1:length(dats)
    x = dats[i][1]
    err = dats[i][2]
    plt = plot!(x, err * (conv_bases[i]^conv_order))
end
return plt
```
